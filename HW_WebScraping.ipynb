```python
from bs4 import BeautifulSoup
import requests

# Send a GET request to get restaurant information from the Yelp website.
url = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc=Irvine%2C+CA&sortby=recommended'

response = requests.get(url)

# If the request is successful (HTTP status code 200)
if response.status_code == 200:
    data = []
    # Parses HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    for div in soup.find_all("div",{"data-testid": "serp-ia-card"}):
        name = div.find("h3").find("a").get_text(strip=True)
        rating = div.find("span", class_="css-gutk1c").get_text(strip=True)
        review_count = div.find("span", class_="css-chan6m").get_text(strip=True).replace("(", "").replace(" reviews)", "")
        #type = div.find("div",class_="css-1kiyre6").find("span", class_="css-11bijt4").get_text(strip=True)
        types = div.find("div", class_="css-1kiyre6").find_all("span", class_="css-11bijt4")
        type_values = [type.get_text(strip=True) for type in types]
        url = div.find("h3").find("a")["href"]

        print(f"Name: {name}\n Rating: {rating}\n Review Count: {review_count}\n Type: {type_values}\n URL: {url}")
        print()

        data.append({
            "name": name,
            "rating": rating,
            "review": review_count,
            "type": type_values,
            "url": url,
        })
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.to_csv("yelp_business_data.csv", index=False)
```
